```python
#Question 2. Python If-Else

if __name__ == '__main__':
    n = int(input().strip())
    if(n%2!=0):
        print("Weird")
    else:
        if n in range(2,6):
            print("Not Weird")
        if n in range(6,21):
            print("Weird")
        if (n >20):
            print("Not Weird")
```

```python
#Question 5. Loops

if __name__ == '__main__':
    n = int(input())
    temp=0
    while(temp!=n):
        print(temp**2)
        temp+=1
```

```python
#Question 6. Write a function

def is_leap(year):
    leap = False

    # Write your logic here
    if(year%4==0 and year%100!=0):
        leap=True
    elif (year%4==0 and year%100==0 and year%400==0):
        leap=True
    return leap

year = int(input())
print(is_leap(year))
```
