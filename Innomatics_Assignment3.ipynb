```python
#Question 1. (1108) Defanging an IP Address
'''Given a valid (IPv4) IP address, return a defanged version of that IP address.
A defanged IP address replaces every period "." with "[.]".'''

class Solution:
    def defangIPaddr(self, address: str) -> str:
        temp=list(address)
        for i in range(len(temp)):
            if temp[i]==".":
                temp[i]="[.]"
        result="".join(temp)
        return result
p1=Solution()
p1.defangIPaddr("1.1.1.1")
```

```python
#Question 2. (1295) Find Numbers with Even Number of Digits
'''Given an array nums of integers, return how many of them contain an even number of digits.'''

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count=0
        for num in nums:
            temp=0
            while (num):
                num=num//10
                temp+=1
            if temp%2==0:
                count=count+1
        return count
nums =[12,345,2,6,7896]
p1=Solution()
p1.findNumbers(nums)
```

```python
#Question 3. (1512) Number of Good Pairs
'''Given an array of integers nums, return the number of good pairs.
A pair (i, j) is called good if nums[i] == nums[j] and i < j.'''

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        count=0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if(nums[i]==nums[j]):
                    count+=1
        return count
nums =[1,2,3,1,1,3]
p1=Solution()
p1.numIdenticalPairs(nums)
```

```python
#Question 4. (1365) How Many Numbers Are Smaller Than the Current Number
'''Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. 
That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].'''

class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        result=[]
        for num in nums:
            count=0
            for i in range(len(nums)):
                if (nums[i]<num):
                    count+=1
            result.append(count)
        return result
nums = [8,1,2,2,3]
p1=Solution()
p1.smallerNumbersThanCurrent(nums)
```

```python
#Question 6. (1486) XOR Operation in an Array
'''You are given an integer n and an integer start.
Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.
Return the bitwise XOR of all elements of nums.'''

class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        nums=[0]*n
        for i in range(n):
            nums[i]=start+2*i
            if (i==0):
                xor=nums[i]
            else:
                xor^=nums[i]
        return xor 
n = 5
start = 0
p1=Solution()
p1.xorOperation(n,start)
```
